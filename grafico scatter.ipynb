```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 20, 200)
y = np.cumsum(np.random.randn(2, 100))
```

```python
fig, ax = plt.subplots(figsize=(12,4))

ax.set_title('gr√°fico', fontsize=14)
plt.plot(x,y, color='grey', lw=2)
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

plt.text(20.2,15.3, 'linha', color='grey',  weight='bold', fontsize=14)
plt.text(7.5,5.6, 'x:7.5, y:9.6',fontsize=10)
ax.scatter(x=7.5, y=9.6, s=1000, color='#F22738')
ax.set_xlabel(None)
#ax.grid()
plt.show()
```
